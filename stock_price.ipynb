```python
#start_index = 1 có nghĩa là cào từ dòng thứ nhất
def crawl_symbols(wd, start_index=1, n_symbol=100):
    web_link = 'https://quotes.vcbs.com.vn/a/exchange.html?symbol=HSX'
    wd.get(web_link)
    time.sleep(3)

    symbol_list = []
    #Lấy table body của table chứa các stock symbol
    tbody = wd.find_element(by=By.XPATH, value='//table[@id="tblScroll1"]/tbody')
    for i in range(start_index, n_symbol+1):
        try:
            symbol = tbody.find_element(by=By.XPATH, value='./tr[{}]/td[contains(@class, "StockSymbol")]'.format(i))
        except NoSuchElementException:
            break
        symbol_list.append(symbol.text)
    return symbol_list
```

## ***Task 1.b - Collecting the daily stock prices***
**Collecting the daily stock prices (closing prices) of those stock symbols in 2020, 2021, 2022**

### Functions

```python
#Hàm tắt quảng cáo nếu có
def close_Ad_if_exists(wd):
    try:
        close_btn = wd.find_element(by=By.XPATH, value='//div[@class="mfp-content"]/*/button[@class="mfp-close"]')
        wd.execute_script("arguments[0].click();", close_btn)
    except NoSuchElementException:
        return False
    return True
```

```python
#Hàm chuyển trang kế tiếp
#Nếu còn trang kế => Chuyển trang và trả về True
#Ngược lại => Trả về False
def click_next_page(wd):
    try:
        #Page navigation để chuyển trang
        page_navigation = wd.find_element(by=By.XPATH, value='//*[@id="ctl00_webPartManager_wp1770166562_wp1427611561_callbackData"]/div[2]')
        #Trang kế tiếp
        next_page = page_navigation.find_element(by=By.XPATH, value='./span/following-sibling::a')
        wd.execute_script("arguments[0].click();", next_page)
    except StaleElementReferenceException:
        #Trang chưa được load hoàn toàn => sleep thêm đợi trang load
        print('StaleElementReferenceException. Trying again after 0.3 seconds...')
        time.sleep(0.3)
    except NoSuchElementException:
        return False
    return True
```

```python
#Hàm cào dữ liệu của 1 stocksymbol nhập vào
#Mặc định là 1 năm, nếu muốn thay đổi khoảng thời gian thì nhập thêm start_date và end_date
def crawl_daily_prices(wd, stocksymbol, list_dates=set(), start_date="", end_date="", sleep_time=0.5):
    web_link = 'https://www.stockbiz.vn/Stocks/' + stocksymbol + '/HistoricalQuotes.aspx'
    wd.get(web_link)
    time.sleep(sleep_time)
    close_Ad_if_exists(wd)
    
    #Chọn ngày tháng bắt đầu, và ngày tháng kết thúc nếu có
    #End date nếu bỏ trống thì mặc định là ngày hiện tại
    if (start_date!=""):
        input_start_date = wd.find_element(by=By.XPATH, value='//input[@id="ctl00_webPartManager_wp1770166562_wp1427611561_dtStartDate_picker_picker"]')
        input_start_date.send_keys(start_date);
    if (end_date!=""):
        input_end_date = wd.find_element(by=By.XPATH, value='//input[@id="ctl00_webPartManager_wp1770166562_wp1427611561_dtEndDate_picker_picker"]')
        input_end_date.send_keys(end_date);
    view_btn = wd.find_element(by=By.XPATH, value='//*[@id="ctl00_webPartManager_wp1770166562_wp1427611561_btnView"]')
    view_btn.click()
    time.sleep(sleep_time)

    closing_price_dict = {}
    while (True):
        try:
            #Table body chứa nội dung cần cào dữ liệu
            tbody = wd.find_element(by=By.XPATH, value='//div[@id="ctl00_webPartManager_wp1770166562_wp1427611561_callbackData"]/table/tbody')

            #Cào từng dòng dữ liệu trong table body
            table_rows = tbody.find_elements(by=By.XPATH, value='./tr')
            for i in range(1, len(table_rows)):
                list_td = table_rows[i].find_elements(by=By.XPATH, value='./td')

                date = list_td[0].text
                closing_price = float(list_td[5].text.replace(',', '.'))*1000
                
                list_dates.add(date)
                closing_price_dict[date] = closing_price
        except StaleElementReferenceException:
            #Trang chưa được load hoàn toàn => sleep thêm đợi trang load
            print('StaleElementReferenceException. Trying again after %.1f seconds...' %sleep_time)
            time.sleep(sleep_time)
            continue

        #Chuyển trang 
        has_next_page = click_next_page(wd) # Sang trang mới
        if not has_next_page:               # Nếu hết trang thì dừng
            break
        time.sleep(sleep_time)              # Chờ trang load
    
    return closing_price_dict, list_dates
```

```python
#Hàm normalize đã được định nghĩa ở task 2
def normalize(row):
    old_price = row.copy()
    normalize = TimeSeriesScalerMeanVariance(mu=0., std=1.)
    new_price = normalize.fit_transform([old_price])
    return old_price, new_price[0]

#Đọc dữ liệu đã cào được
df = pd.read_csv("allDailyStockPrice.csv", index_col = 0)
df
```

## Functions

```python
#Lấy daily stock price của stock symbol nhập vào
def get_daily_stock_price(stock_sym):
    stock_data = df.loc[stock_sym].values
    dates = df.columns
    #Chuẩn hóa
    original, normalized = normalize(stock_data)
    #Đổi shape cho phù hợp để đưa vào mô hình huấn luyện
    normalized = normalized.reshape(len(normalized), 1)
    dates = np.array(dates).reshape(len(dates), 1)
    return original, normalized, dates

#Chia tập data thành tập train và test
#- Train: năm 2020 và 2021
#- Test: từ năm 2022 đến hiện tại
def train_test_split(input_dates, input_prices):
    for i in range(input_dates.shape[0]):
        year = input_dates[i,0].split('/')[2]
        if year=='2022':
            break
    train_price = input_prices[:i, :]
    test_price = input_prices[i:, :]

    train_date = input_dates[:i, :]
    test_date = input_dates[i:, :]
    
    return train_price, test_price, train_date, test_date

#Chia đầu vào X, đầu ra y
#Giá trị look_back bằng x nghĩa là chúng ta lấy x bước thời gian trước, để dự đoán cho thời gian tiếp theo.
def X_y_split(dataset, look_back):
    X, y = [], []
    for i in range(look_back, len(dataset)):
        X.append(dataset[(i-look_back):i, :])
        y.append(dataset[i:(i+1), 0])
    return np.array(X), np.array(y)
```

```python
def create_and_train_model(X_train, y_train, X_test, y_test, epochs=50, batch_size=20, verbose=1):
    model = Sequential()
    
    model.add(LSTM(50, activation='relu',input_shape=(X_train.shape[1], X_train.shape[2]), return_sequences=True))
    model.add(Dropout(rate=0.1))
    
    model.add(LSTM(50, activation='relu', return_sequences=True))
    model.add(Dropout(rate=0.1))
    
    model.add(LSTM(50, activation='relu', return_sequences=False))
    model.add(Dropout(rate=0.1))
    
    model.add(Dense(1))
    model.summary()
    
    model.compile(loss = 'mean_squared_error', optimizer = 'adam')
    history = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs = epochs, batch_size = batch_size, verbose=verbose)
    return model
```

```python
def predictPrice(model, train_data, look_back, end_date=datetime(2022, 4, 30)):
    #Nhập train_data để lấy dữ liệu x ngày cuối của năm 2021, dự đoán cho ngày đầu của năm 2022
    start_date=datetime(2022, 1, 1)
    dates = pd.date_range(start=start_date, end=end_date)
    dates_str = [d.strftime('%d/%m/%Y') for d in dates]
    
    num_days = len(dates_str)
    #look_back là x, có nghĩa là Lấy dữ liệu x ngày cuối năm 2021 để dự đoán ngày đầu năm 2022
    X_predict = np.zeros((1, num_days+look_back, 1))
    X_predict[0, 0:look_back] = train_data[-look_back:]

    #Nếu không nhập end_date, mặc định dự đoán 4 tháng đầu năm 2022
    y_pred=np.zeros((num_days, 1))
    for d in range(num_days):
        new_price = model.predict(X_predict[:, d:d+look_back, :])[0]
        X_predict[0][d+look_back] = new_price
        y_pred[d,:] = new_price
    
    return y_pred, np.array(dates_str)
```

```python
def inverse_normalize(org_data, normalized_pred):
    #org_data là dữ liệu huấn luyện gốc trước khi chuẩn hóa
    normalized_pred = normalized_pred.ravel()
    mean = np.sum(org_data)/org_data.shape[0]
    sum_val = 0.0
    for x in org_data:
        sum_val += (x - mean)** 2
    std = math.sqrt(sum_val/org_data.shape[0])
    
    x = [round((y*std) + mean) for y in normalized_pred]
    return x
```

```python
def visualizePrediction(known_dates, known_price, predict_dates, predicted_price, symbol, title='STOCK PRICE PREDICTION'):
    #Chuẩn bị dữ liệu plot x-axis
    known_dates = known_dates.ravel()
    predict_dates = predict_dates.ravel()
    
    start_date = datetime.strptime(known_dates[0], '%d/%m/%Y')
    end_date = datetime.strptime(known_dates[-1], '%d/%m/%Y')
    d = datetime.strptime(predict_dates[-1], '%d/%m/%Y')
    if end_date<d:
        end_date = d
    all_dates = [d.strftime('%d/%m/%Y') for d in pd.date_range(start=start_date, end=end_date)]
    
    #Đánh số các ngày
    all_dates_index = [i for i in range(len(all_dates))]
    known_dates_index = [all_dates.index(d) for d in known_dates]
    predict_dates_index = [all_dates.index(d) for d in predict_dates]
    
    #Vẽ hình
    fig = plt.figure(figsize=(15, 8))
    plt.plot(known_dates_index, known_price, color='blue', linewidth=3, label='Actual_Data')
    plt.plot(predict_dates_index, predicted_price, color='orange', linewidth=3, label='Predicted_Price')
    #Vẽ x-axis (ngày tháng năm)
    ax = plt.gca()
    ax.xaxis.set_ticks([all_dates_index[i] for i in range(1, len(all_dates_index), len(all_dates_index)//10)])
    ax.xaxis.set_major_formatter(FuncFormatter(lambda x, y: format(all_dates[x])))
    #Gán nhãn x-axis, y-axis
    plt.title('%s - SYMBOL: %s' %(title, symbol), fontsize='18')
    plt.xlabel('Dates', fontsize=13)
    plt.ylabel('Stock Price', fontsize=13)
    
    plt.grid()
    plt.legend(loc='upper left', prop={'size': 12})
    plt.show()
```
